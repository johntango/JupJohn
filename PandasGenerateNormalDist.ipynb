# Normal Distribution Probability Density Function
$
\text{A probability density function (PDF) tells us the probability of a variable $ x$ lying between two values.}\\ 
\text{The area of the PDF between those two values is that probability} \\
$
$
\text{Probability density function (PDF) of the normal distribution is} \\
$
$$
f(x)=\frac{1}{\sigma \sqrt{2 \pi}}e^{-\frac{(x- \mu)^2}{2 \sigma^2}}
$$

The random variable $X$ described by the PDF is a normal variable that follows a normal distribution with mean $\mu$ and variance $\sigma^2$.

$
\text{Normal distribution notation is} \\
$
$$
X \sim N(\mu,\sigma^2) \\
$$

The total area under the PDF curve equals 1.

## Covariance and Correlation Coefficients

### Mean or Expected Value
$$ \mu_{x} = \frac{\sum_{n} x_n } {N}$$
$$ \mu_{y} = \frac{\sum_{n} y_n } {N}$$
#### 
$$ X = x -\mu_{x} $$
$$ Y = y -\mu_{y} $$


### Covariance of two variables - Joint variability of two random variables
$$cov(x,y) =  \frac{\sum_{n} X_n * Y_n} {N}$$
#### Note that the values can vary from $+\infty $ to $ -\infty  $
#### The special case of variance of a variable with itself
$$cov(x,x) =  \frac{\sum_{n} X_n * X_n} {N} = \sigma(x)^{2}= \sigma_{x}^{2}  $$

```python
# Covariance Calculation  - results can vary from +infinity to -infinity
import numpy as np
import matplotlib.pyplot as plt
x = np.array([-1,0.,1])
y = np.array([-100,0.,100])

plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
cov = np.sum(x*y)/3
cov
```

### Pearson Correlation, denoted by r or R

$$r =  \frac{1}{N} \sum_{n} \frac{X_n}{\sigma_x} *  \frac{Y_n}{\sigma_y}$$
or 
$$r =  \frac{1}{N} \frac{\sum_{n} X_n *  Y_n}{\sigma_x *\sigma_y}$$
this can be written as 
$$r =  \frac{\sum_{n} X_n * Y_n} {\sqrt((\sum_{n}X_n*X_n) * (\sum_{n} Y_n * Y_n))}$$
we note that 
$$ \sigma_x = \sqrt{ \frac{\sum(x -\mu_{x})*(x -\mu_{x})}{N} } = \sqrt{\frac{\sum X*X}{N}}$$
#### Note that the values range between +1 and -1

### Difference between correlation and covariance
#### Think of correlation as "normalized" covariance

$$corr = \frac{\sigma_{xy}}{\sigma_{x}*\sigma_{y}}$$
$$corr =  \frac{cov(x * y)}{\sigma_{x}*\sigma_{y}} =  \frac{\sum_{n} X_n * Y_n} {N*\sqrt{\sigma_{x}^{2}*\sigma_{y}^{2}}}$$

```python
# Pearson Correlation Calculation
import numpy as np
x = np.array([-1,0.,1])
y = np.array([-10000,0.,10000])
cov = np.sum(x*y)/3

xy = (x[0]*y[0] + x[1]*y[1] + x[2]*y[2])/3
xx = (x[0]*x[0] + x[1]*x[1] + x[2]*x[2])/3
yy = (y[0]*y[0] + y[1]*y[1] + y[2]*y[2])/3
xy = xy/(np.sqrt(xx * yy))
xy
```

## Calculate Difference between Sample and Population Stats

```python
# Generate a Population and calculate its mean and variance
import matplotlib.pyplot as plt
from scipy.stats import norm
Npoints = 200
diffMean = []
diffVariance = []
bestMean = []
bestVar = []
p = np.random.uniform(0,100,Npoints)
pmean = p.mean()
pvar = np.var(p)
# take a sample and calc mean and variance
N = 1000

for i in range(N):
    sample = np.random.choice(p,50) # choose a sample from population
    smean = sample.mean()           # sample mean
    svar = np.var(sample,ddof=1)    # sample variance
                                    # ddof= 0 or 1 means divide by N or N-1
    diffMean.append(pmean-smean)    # mean difference between sample and population 
    bestMean.append(np.sum(diffMean)/(len(diffMean))) # accumulate and average
    diffVariance.append(pvar-svar)  # variance difference between sample and population
    bestVar.append(np.sum(diffVariance)/len(diffVariance)) # accumulate and average
    
# plot results


plt.scatter(range(N),bestMean,color='b', label='Mean Error')
plt.ylabel('Error')
plt.scatter(range(N),bestVar,color='g', label='Variance Error')
plt.legend(loc='upper right')
print(pmean, pvar)
```


Polulation Variance $$ \sigma^{2} = \frac{\sum(x -\mu_{x})*(x -\mu_{x})}{N}  $$

Sample Variance $$ s^{2} = \frac{\sum(x -\mu_{x})*(x -\mu_{x})}{N-1}  $$









```python
population = np.array([5, 4, 4, 9, 9, 8, 2, 0, 3, 6])
sample = np.array([5, 4, 4, 8, 6])

def variance(sample, mean, bias):
    N = len(sample)
    s2 = 0
    for i in range(0,N):
        x = sample[i]-mean
        s2 = s2 + (x * x) 
    var = s2/(N - bias)
    return var

N = 10

var_pop = []
var_sample = []

for mean in range(0,N):   # shift the mean from 0 to N
    var_pop.append( variance(population, mean, 0))

for mean in range(0,N):   # shift the mean from 0 to N
    var_sample.append( variance(sample, mean, 1))
# Plotting
plt.scatter(range(N),var_pop,color='b', label='Population')

plt.scatter(range(N),var_sample,color='r', label='Sample')
plt.ylim([0, 50])
plt.ylabel('Variance')
plt.xlabel('Mean')
plt.legend(loc='upper right')
```
